```python
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get("http://pythonhow.com/example.html")
type(r)
```

```python
c=r.content
c
```

```python
soup=BeautifulSoup(c,"html.parser")
print(soup.prettify())
```

```python
all=soup.find_all("div",{"class":"cities"})
one=soup.find("div",{"class":"cities"})#finds the first item
print(all)
```

```python
type(all)#a special beautiful string, BS will know where the tags are.
```

```python
#this object supports indexing:
all=soup.find_all("div",{"class":"cities"})
print(all)
```

```python
all[0].find_all("h2")
```

```python
all[0].find_all("h2")[0].text
```

```python
for item in all:
    print(item.find_all("h2")[0].text)
```
